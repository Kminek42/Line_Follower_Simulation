```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
import re
```

```python
output_path = '../data/results'
```

```python
results = {}
pattern = "..[[].*[]]"

for filename in os.listdir(output_path):
    model_name = re.search(pattern, filename).group()
    results[model_name] = np.loadtxt(f'{output_path}/{filename}', delimiter=",").tolist()

results = pd.DataFrame(results)
results.index *= 10
```

```python
results
```

```python
plt.rcParams["figure.dpi"] = 140

palette = sns.color_palette("tab20", n_colors=len(results.columns))
results.plot(
    figsize=(12, 8),
    color=palette,
    linewidth=2,
    legend=True,
    title="Oceny modeli dla rÃ³Å¼nych parametrÃ³w nauki",
    xlabel='ÅÄ…czna liczba osobnikÃ³w',
    ylabel='Åšrednia prÄ™dkoÅ›Ä‡ [m/s]',
    
)
plt.show()
```
