```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def add_rectangle(ax, center, width, height, **kwargs):
    """
    Add a rectangle to the given Axes object.

    Parameters:
        ax (matplotlib.axes.Axes): The axes to add the rectangle to.
        center (tuple): The (x, y) coordinates of the rectangle's center.
        width (float): The width of the rectangle.
        height (float): The height of the rectangle.
        **kwargs: Additional keyword arguments for the rectangle's styling
                  (e.g., edgecolor, facecolor, linewidth).
    """
    x, y = center
    bottom_left = (x - width / 2, y - height / 2)
    rect = patches.Rectangle(bottom_left, width, height, **kwargs)
    ax.add_patch(rect)
```

```python
fig, ax = plt.subplots(dpi=200)
ax.set_xlim(-1, 1)
ax.set_ylim(-0.5, 0.5)
ax.set_aspect('equal')

add_rectangle(ax, center=(-0.7, -0.05), width=0.5, height=0.3, edgecolor='red', facecolor='red', linewidth=2, alpha=0.8)
add_rectangle(ax, center=(-0.65, 0.0), width=0.5, height=0.3, edgecolor='red', facecolor='red', linewidth=2, alpha=0.4)
add_rectangle(ax, center=(-0.6, 0.05), width=0.5, height=0.3, edgecolor='red', facecolor='red', linewidth=2, alpha=0.2)

add_rectangle(ax, center=(-0.1, -0.05), width=0.1, height=0.5, edgecolor='blue', facecolor='blue', linewidth=2, alpha=0.8)
add_rectangle(ax, center=(-0.05, 0.0), width=0.1, height=0.5, edgecolor='blue', facecolor='blue', linewidth=2, alpha=0.4)
add_rectangle(ax, center=(-0.0, 0.05), width=0.1, height=0.5, edgecolor='blue', facecolor='blue', linewidth=2, alpha=0.2)

add_rectangle(ax, center=(0.3, -0.05), width=0.1, height=0.3, edgecolor='green', facecolor='green', linewidth=2, alpha=0.8)
add_rectangle(ax, center=(0.35, 0.0), width=0.1, height=0.3, edgecolor='green', facecolor='green', linewidth=2, alpha=0.4)
add_rectangle(ax, center=(0.4, 0.05), width=0.1, height=0.3, edgecolor='green', facecolor='green', linewidth=2, alpha=0.2)



ax.plot([-0.3, -0.2], [-0.05, 0.05], color='grey', linestyle='-', linewidth=2)
ax.plot([-0.3, -0.2], [0.05, -0.05], color='grey', linestyle='-', linewidth=2)

ax.plot([0.1, 0.2], [-0.02, -0.02], color='grey', linestyle='-', linewidth=2)
ax.plot([0.1, 0.2], [0.02, 0.02], color='grey', linestyle='-', linewidth=2)


plt.show()
```
